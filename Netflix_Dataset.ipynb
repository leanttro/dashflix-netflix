```python
import pandas as pd
import numpy as np
import sqlite3
from sqlalchemy import create_engine
```

```python
df_netflix = pd.read_csv("Netflix Dataset.csv")
```

```python
df_netflix.head()
```

```python
df_netflix.describe()
```

```python
engine = create_engine("sqlite:///lakehouse.db")
df_netflix.to_sql("df_netlix_bronze", engine, if_exists="replace", index=False)
```

```python
df_netflix_silver = df_netflix.drop_duplicates()
```

```python
df_netflix_silver
```

```python
df_netflix_silver.to_sql('df_netflix_gold', con=engine, if_exists='replace', index=False)
```

```python
df_netflix_silver.to_csv('netflix_gold.csv', index=False)
```
